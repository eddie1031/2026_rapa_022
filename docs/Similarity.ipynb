```python
from langchain_ollama.embeddings import OllamaEmbeddings

embedding = OllamaEmbeddings(model='qwen3-embedding:latest')
```

```python
import numpy as np

def cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    return float(dot_product / (norm_a * norm_b))

```

```python
APPLE = "사과 과일"
PEAR = "배 과일"
COMPUTER = "데스크톱 컴퓨터"
MONITOR = "컴퓨터 모니터"
```

```python
doc = [ APPLE, PEAR, COMPUTER, MONITOR ]

vectors = embedding.embed_documents(doc)
```

```python
apple_pear_result = cosine_similarity(vectors[0], vectors[1])
apple_computer_result = cosine_similarity(vectors[0], vectors[2])
apple_monitor_result = cosine_similarity(vectors[0], vectors[3])

computer_monitor_result = cosine_similarity(vectors[2], vectors[3])

print(f'apple_pear_result = {apple_pear_result:.2f}')
print(f'apple_computer_result = {apple_computer_result:.2f}')
print(f'apple_monitor_result = {apple_monitor_result:.2f}')
print(f'computer_monitor_result = {computer_monitor_result:.2f}')
```
