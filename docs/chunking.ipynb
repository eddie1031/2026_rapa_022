```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

text_loader = TextLoader(
    file_path='../libs/split_text.txt',
    encoding='utf-8',
)

doc = text_loader.load()[0]

splitter = CharacterTextSplitter(
    separator='\n',
    chunk_size=30,
    chunk_overlap=20,
)

chunk_lines = splitter.split_text(doc.page_content)

print(chunk_lines)

```
