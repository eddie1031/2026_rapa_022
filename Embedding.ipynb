```python
from langchain_ollama.embeddings import OllamaEmbeddings

embedding = OllamaEmbeddings(model='qwen3-embedding:latest')
```

```python
apple = "사과"

vector = embedding.embed_query(apple)

print(vector)
print(len(vector))
```

```python
docs = [ '안녕하세요!', '반갑습니다!' ]

vectors = embedding.embed_documents(docs)

print(vectors[0][1:5], vectors[1][1:5])
```
